## Perceptron: Caso 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados
np.random.seed(42)  # Estamos controlando a aleatoriedade em seus cálculos
n_samples = 50
X_red = np.random.randn(n_samples, 2) + np.array([-2, -2])
X_black = np.random.randn(n_samples, 2) + np.array([2, 2])

# Rotulos: 1 para vermelho, -1 para preto
y_red = np.ones(n_samples)
y_black = -np.ones(n_samples)

# Combina os dados
X = np.vstack((X_red, X_black))
y = np.hstack((y_red, y_black))

# Função para plotar os dados antes do treinamento
def plot_data(X, y):
    plt.figure(figsize=(10, 6))
    for i in range(X.shape[0]):
        if y[i] == 1:
            plt.scatter(X[i, 0], X[i, 1], color='red')
        else:
            plt.scatter(X[i, 0], X[i, 1], color='black')

    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Dados do Treinamento do Perceptron')
    plt.grid(True)
    plt.show()

# Plotar os dados antes do treinamento
plot_data(X, y)
```

```python
# Adiciona uma coluna de 1s para o bias
X_bias = np.hstack((X, np.ones((X.shape[0], 1))))

# Função de treinamento do perceptron
def train_perceptron(X, y, lr=0.01, epochs=1000):
    weights = np.zeros(X.shape[1])
    for epoch in range(epochs):
        for i in range(X.shape[0]):
            # y_final = sign(∑x_i*w_i)
            y_pred = np.sign(np.dot(X[i], weights))
            if y[i] * y_pred <= 0:
                weights += lr * y[i] * X[i]
    return weights

# Treina o perceptron
weights = train_perceptron(X_bias, y)

# Função para plotar os dados e a linha de decisão
def plot_perceptron(X, y, weights):
    plt.figure(figsize=(10, 6))
    for i in range(X.shape[0]):
        if y[i] == 1:
            plt.scatter(X[i, 0], X[i, 1], color='red')
        else:
            plt.scatter(X[i, 0], X[i, 1], color='black')

    # Equação da linha de decisão: w1*x1 + w2*x2 + w3 = 0
    x1 = np.array([min(X[:, 0]), max(X[:, 0])])
    x2 = -(weights[0] * x1 + weights[2]) / weights[1]

    plt.plot(x1, x2, color='blue', label='Linha de Decisão')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Perceptron em R^2 para Separar Classes')
    plt.legend()
    plt.grid(True)
    plt.show()


# Plotar os dados e a linha de decisão do perceptron após o treinamento
plot_perceptron(X, y, weights)
```

## Caso Geral

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import Perceptron
from sklearn.datasets import make_circles

# Gerar dados para o problema com círculos concêntricos
np.random.seed(42)
X, y = make_circles(n_samples=100, noise=0.1, factor=0.5)

# Criar e treinar o Perceptron
perceptron = Perceptron(max_iter=1000, random_state=42)
perceptron.fit(X, y)

# Criar e treinar o MLP
mlp = MLPClassifier(hidden_layer_sizes=(5,), max_iter=1000,
                    activation='relu', solver='adam', random_state=42)
mlp.fit(X, y)

# Função para plotar os dados e a fronteira de decisão
def plot_decision_boundary(X, y, model, title):
    plt.figure(figsize=(10, 6))
    h = .02  # passo da malha de pontos
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.RdYlBu)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Plotar a fronteira de decisão do Perceptron
plot_decision_boundary(X, y, perceptron,
                       'Perceptron - Dados com Círculos Concêntricos')

# Plotar a fronteira de decisão do MLP
plot_decision_boundary(X, y, mlp, 'MLP - Dados com Círculos Concêntricos')
```
